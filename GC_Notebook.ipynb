```python
#imports
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
```

```python
#file path collection
from google.colab import drive
drive.mount('/content/drive')

# Path to the main folder
main_folder = '/content/drive/My Drive/for_Andrei'

# List all files and subfolders in the main folder
files_list = []
unique_letters = set()
unique_letter_positions = []
for i, file_path in enumerate(os.listdir(main_folder)):
    if os.path.isfile(os.path.join(main_folder, file_path)):
      files_list.append(file_path)
      first_letter = file_path[0].lower()  # Convert to lowercase for case-insensitivity
      if first_letter not in unique_letters:
        unique_letters.add(first_letter)
        unique_letter_positions.append(i)

```

```python
print(len(files_list))
print(len(unique_letters))
print(unique_letters)
print(unique_letter_positions)
unique_file_names = []
for position in unique_letter_positions:
  unique_file_names.append(files_list[position])
print(unique_file_names)
```

```python
lat_name = 'lat'
lat_file_path = os.path.join(main_folder, lat_name)
full_summary_name = "full_summary.csv"
full_summary_path = os.path.join(main_folder, full_summary_name)
hydrogen_name = "H_"
hydrogen_file_path_list = []
nitrogen_name = "N_"
nitrogen_file_path_list = []

for file_name in files_list:
  if file_name.startswith(hydrogen_name):
    hydrogen_file_path_list.append(os.path.join(main_folder, file_name))
  if file_name.startswith(nitrogen_name):
    nitrogen_file_path_list.append(os.path.join(main_folder, file_name))

print(len(hydrogen_file_path_list))
print(len(nitrogen_file_path_list))
```

```python
temp_path = hydrogen_file_path_list[0]
with open(temp_path, 'r') as file:
  lines = file.readlines()
  print(lines)
  print(temp_path)
  lines = [float(number) for line in lines for number in line.strip().split()[:3]]
  print(lines)
  nested_lines = []
  for i in range(len(lines)):
    if i % 3 == 0:
      nested_lines.append(lines[i:i+3])
  print(nested_lines)
  array = np.array(nested_lines)
  print(array)
```

```python
import time

start_time_1 = time.time()
main_data = {"Hydrogen": {}, "Nitrogen": {}}
for i, file_path in enumerate(hydrogen_file_path_list):
  with open(file_path, 'r') as file:
    path_name = os.path.basename(file_path)
    lines = file.readlines()
    lines = [float(number) for line in lines for number in line.strip().split()[:3]]
    nested_lines = []
    for i in range(0, len(lines), 3):
      try:
        nested_lines.append(lines[i:i+3])
      except IndexError:
        print(i)
        print(len(lines))
        print(f"Warning: Incomplete data in file {path_name}, skipping last line.")
    print(nested_lines)
    array = np.array(nested_lines)
    main_data["Hydrogen"][path_name] = array
end_time_1 = time.time()
print(end_time_1 - start_time_1)

start_time_2 = time.time()
for i, file_path in enumerate(nitrogen_file_path_list):
  with open(file_path, 'r') as file:
    path_name = os.path.basename(file_path)
    lines = file.readlines()
    lines = [float(number) for line in lines for number in line.strip().split()[:3]]
    nested_lines = []
    for i in range(0, len(lines), 3):
      try:
        nested_lines.append(lines[i:i+3])
      except IndexError:
        print(i)
        print(len(lines))
        print(f"Warning: Incomplete data in file {path_name}, skipping last line.")
    print(nested_lines)
    array = np.array(nested_lines)
    main_data["Nitrogen"][path_name] = array

print(main_data)
end_time_2 = time.time()
print(end_time_2 - start_time_2)
```
